# 3) Graphs

## Histogram

```python
# Hist graph to see the distribution of the values in the columns
df.hist(figsize=(14, 8), bins=20)

plt.tight_layout()
plt.show()
```

### Notes of histogram

#### Normal variables
- Subject: Only to identify the pacient
- Age: The most part is in 55-75 years (the normal for Parkinson)
- Sex: 0 = Male, 1 = Female (The most part is Men)
- test_time: Number of days since recruitment (distribueted)

#### UPDRS (Unified parkinson's Disease Rating Scale)
- Scale for gravity of parkinson (The bigger, the worse)

- In our studies, we have motor distribueted and total with center concentration

- Motor_updrs: Motor symptoms (tremor, rigidity, difficulty walking, etc.)
- total_updrs: Motor symptoms and others from the desease

#### Voice columns
- The research aims to predict Parkinson's disease using voice characteristics.

##### Jitter
- Voice frequency variation (the bigger, the instability is major)
- This graphs don't have a lot of variation

- This is 5 ways of make the same measure

- jitter: normal person (aaaaaa) parkinson person (aaAaaAaaA)
- jitter_abs: same of jitter, but in seconds
- jitter_rap: relative average perturbation 
- jitter_ppq5: Pitch Period Perturbation Quotient
- jitter_ddp: Another way of measure

##### Shimmer
- Voice amplitude
- We don't have a lot of shimmer too

- Ways of do the same measure

- shimmer: normal person (AAAAAAAAAAAA), parkinson person (AAAaaAaAaAAaaA)
- shimmer_db
- shimmer_apq3
- shimmer_apq5
- shimmer_apq11
- shimmer_dda

##### nhr
- Noise-to-Harmonics Ratio (amount of noise in the voice)
- The bigger, the worse

##### hnr
- Harmonics-to-Noise Ratio (Inverse of nhr)

##### rpde
- Recurrence Period Density Entropy (how predictable or irregular the voice pattern is)
- The bigger, the worse

##### dfa
- Detrended Fluctuation Analysis (Voice correlation in long term)

##### ppe
- Pitch Period Entropy (Voice irregular tone)
- The bigger, more desorganization the voice is

## Correlation

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12, 10))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.tight_layout()
plt.show()
```

### Notes

#### Motor UPDRS and Total UPDRS
- 0.95 = The two have the "same porpuse", so if we will use, we need to use only one of them

#### Jitter
- 0.98 - 1.00 = All of them have the same objective, we only need one a time

#### Shimmer
- 0.98 - 1.00 = All of them have the same objective, we only need one a time

#### HNR
- negative correlation with the Jitter, shimmer and ppe
- HNR is low when we have instability (AAaaaAaAAAa) and Jitter, shimmer is bigger when we have instability, so, they are contrary

#### NHR
- Contrary of HNR

#### Age, sex and test time
- less correlation

#### Subject
- not considered


#### Groups
- Frequency (Jitter)
- Intensity (shimmer)
- Harmonics (nhr, hnr)
