# 1) Import Libraries

```python
# Initial configs
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

# 2) Load Dataset

```python
df = pd.read_csv(r"..\data\dataset.csv")
```

```python
df.head()
```

# 3) Remove Unnecessary Columns

```python
# Column for subject identity, not useful for our ML
# We will use the motor_updrs, so the total will not be useful
df = df.drop(columns=["subject", "total_updrs"])
df.head()
```

We can run some tests with columns such as test_time, Jitter, and Shimmer, and observe which one the model responds to best.

# 4) Handle Categorical Variables

```python
# Transforming sex in 0/1
df["sex"] = df["sex"].astype(int)
```

# 5) Define Target

```python
X = df.drop(columns=["motor_updrs"])
y = df['motor_updrs']

print(X.head())
print("----------------------------------------------------------------------")
print(y.head())
```

# 6) Feature Selection

Here we need to decide if we will use all of the variables that we have, or if we should use only the principals.

In my case, and using the EDA stage, we can choose the second way and use the variables that make more sense.

```python
# Using correlation, this is the best test (but we can change for shimmer)
X = X[["age", "sex", "jitter", "shimmer_apq11", "hnr", "rpde", "dfa", "ppe"]]

X.head()
```

# 7) Train/Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

# 8) Scaler Variables

```python
# Doing the scalar of the variables, because we need all in the same scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

# 9) Save Scaler and train/test

```python
import joblib
import os

os.makedirs("../models", exist_ok=True)

joblib.dump(scaler, r"../models/scaler.pkl")
```

```python
import os

os.makedirs("../data/processed_data", exist_ok=True)


pd.DataFrame(X_train_scaled).to_csv(
    "../data/processed_data/X_train.csv",
    index=False
)

pd.DataFrame(X_test_scaled).to_csv(
    "../data/processed_data/X_test.csv",
    index=False
)

pd.DataFrame(y_train).to_csv(
    "../data/processed_data/y_train.csv",
    index=False
)

pd.DataFrame(y_test).to_csv(
    "../data/processed_data/y_test.csv",
    index=False
)
```
